**Q1. What is hypothesis testing in statistics?**

**Ans.** Hypothesis testing in statistics is a method used to make decisions or inferences about population parameters based on sample data. It involves evaluating two competing statements (hypotheses) about a population to determine which one is more consistent with the observed data.

**Key Components of Hypothesis Testing:**

**1. Null Hypothesis (H‚ÇÄ):**

- A default assumption or claim.

- Typically states that there is no effect, no difference, or no relationship.

- Example: H‚ÇÄ: Œº = 50 (the population mean is 50)

**2.Alternative Hypothesis (H‚ÇÅ or H‚Çê):**

- The statement you are trying to find evidence for.

- Represents a new claim or effect.

- Example: H‚ÇÅ: Œº ‚â† 50 (the population mean is not 50)

**3.Significance Level (Œ±):**

- The probability of rejecting the null hypothesis when it is actually true (Type I error).

- Common values: 0.05, 0.01

**4. Test Statistic:**

- A standardized value (e.g., z-score, t-score) calculated from sample data used to make a decision about the hypotheses.

**5. P-value:**

- The probability of observing the test statistic or something more extreme, assuming H‚ÇÄ is true.

- If the p-value ‚â§ Œ± ‚Üí reject H‚ÇÄ

- If the p-value > Œ± ‚Üí fail to reject H‚ÇÄ

**6. Decision:**

- Based on the p-value or test statistic compared to critical values.

- You either reject H‚ÇÄ or fail to reject H‚ÇÄ (you never "accept" H‚ÇÄ).



**Q2. What is the null hypothesis, and how does it differ from the alternative hypothesis?**

**Ans.** The null hypothesis (H‚ÇÄ) and the alternative hypothesis (H‚ÇÅ or H‚Çê) are the two competing statements in hypothesis testing. Here's a clear breakdown of each and how they differ:

**Null Hypothesis (H‚ÇÄ)**
**Definition:** A statement that there is no effect, no difference, or no relationship.

**Role:** Serves as the default or status quo assumption.

**Purpose:** It's the hypothesis that you try to disprove or reject using statistical evidence.

**Examples:**
- H‚ÇÄ: Œº = 100 (the population mean is 100)

- H‚ÇÄ: p‚ÇÅ = p‚ÇÇ (proportions in two groups are equal)

- H‚ÇÄ: There is no association between two variables

**Alternative Hypothesis (H‚ÇÅ or H‚Çê)**
**Definition:** A statement that contradicts the null hypothesis.

**Role:** Represents what you aim to support or find evidence for.

**Purpose:** Suggests there is an effect, difference, or relationship.

**Examples:**

- H‚ÇÅ: Œº ‚â† 100 (the mean is not 100)

- H‚ÇÅ: p‚ÇÅ ‚â† p‚ÇÇ (the two proportions are different)

- H‚ÇÅ: There is an association between two variables

**Key Differences:-**

| Feature                                        | Null Hypothesis (H‚ÇÄ)              | Alternative Hypothesis (H‚ÇÅ)       |
| ---------------------------------------------- | --------------------------------- | --------------------------------- |
| Definition                                     | No effect/difference/relationship | There **is** an effect/difference |
| Assumed true until evidence suggests otherwise | ‚úÖ Yes                             | ‚ùå No                              |
| What you test directly                         | ‚úÖ Tested using data               | Inferred if H‚ÇÄ is rejected        |

**Q3. What is the significance level in hypothesis testing, and why is it important?**

**Ans.** The significance level, denoted by Œ± (alpha), is the threshold probability used to decide whether to reject the null hypothesis (H‚ÇÄ).

It is the maximum probability of making a Type I error, which is rejecting the null hypothesis when it is actually true.

**Common choices:**

- Œ± = 0.05 (5% chance of Type I error)

- Œ± = 0.01 (1%)

- Œ± = 0.10 (10%)

**1.Controls Risk of False Positives:**

- By setting Œ±, you decide how much risk you're willing to take in falsely rejecting H‚ÇÄ.

- Example: With Œ± = 0.05, you accept a 5% risk of claiming a significant result when none exists.

**2.Determines the Rejection Region:**

- Helps define the critical values or cutoff for test statistics.

- If the p-value ‚â§ Œ±, you reject H‚ÇÄ.

**3. Affects the Strength of Evidence Needed:**

- A lower Œ± (e.g., 0.01) requires stronger evidence to reject H‚ÇÄ.

- A higher Œ± (e.g., 0.10) is more lenient.

**4. Guides Scientific Rigor:**

- In fields like medicine, a low Œ± (e.g., 0.01) is used to reduce the chance of false claims.

- In exploratory research, a higher Œ± (e.g., 0.10) might be acceptable.

The p-value is the probability of observing a test statistic as extreme as (or more extreme than) the one calculated from your sample data, assuming the null hypothesis is true.

**Q6. What are Type 1 and Type 2 errors in hypothesis testing?**

**Ans.**  In hypothesis testing, you're making a decision based on data ‚Äî but there‚Äôs always a risk of making a wrong decision. These wrong decisions are classified as Type I and Type II errors.

**Type I Error (False Positive)**
**Definition:** Rejecting the null hypothesis (H‚ÇÄ) when it is actually true.

**Interpretation:** You think there is an effect or difference, but there isn‚Äôt.

**Symbol:** Œ± (alpha) ‚Äî the significance level

**Example:** A COVID test says someone is positive (infected), but they are actually healthy.

**Type II Error (False Negative)**

**Definition:** Failing to reject the null hypothesis when the alternative hypothesis (H‚ÇÅ) is actually true.

**Interpretation:** You miss a real effect or difference.

**Symbol:** Œ≤ (beta)

**Power of a test:** Defined as 1 ‚àí Œ≤, the probability of correctly detecting a true effect.

**Example:** A cancer test fails to detect cancer in someone who actually has it.

**Q14. What is a confidence interval, and how is it used to interpret statistical results?**

**Ans.** A confidence interval is a range of values derived from a sample that is likely to contain the true population parameter (such as the mean or proportion) with a certain level of confidence (e.g., 95%).

A confidence interval gives you a range of plausible values for an unknown parameter and reflects the uncertainty inherent in using a sample to estimate a population value.

**Confidence Intervals Are Important:**

| Purpose                               | Explanation                                                                                                           |
| ------------------------------------- | --------------------------------------------------------------------------------------------------------------------- |
| **Estimation**                        | CI provides a more complete picture than a point estimate alone                                                       |
| **Uncertainty quantification**        | Shows how precise your estimate is                                                                                    |
| **Decision-making**                   | If the CI **excludes a value of interest** (e.g., 0 or a hypothesized mean), it suggests **statistical significance** |
| **Alternative to hypothesis testing** | CI can replace p-values in some contexts for interpreting effect size and significance                                |

**Q21. What are the different types of ANOVA tests?**

**Ans.** Types of ANOVA (Analysis of Variance) Tests
ANOVA comes in several forms depending on the number of independent variables and the design of the study. Here‚Äôs a breakdown of the most common types:

**1. One-Way ANOVA**

- Use: Compares the means of 3 or more groups based on 1 independent variable.

- Example: Comparing average test scores of students from 4 different schools.

**2. Two-Way ANOVA**

- Use: Examines the effect of 2 independent variables on a dependent variable.

- Can also detect if there is an interaction effect between the two factors.

**3. Repeated Measures ANOVA**

- Use: Used when the same subjects are measured multiple times (within-subjects design).

- Helps control for individual variability.

**4. MANOVA (Multivariate ANOVA)**

- Use: Like ANOVA, but for multiple dependent variables simultaneously.

- Tests whether groups differ across a set of related outcome variables.

**5. ANCOVA (Analysis of Covariance)**

- Use: Similar to ANOVA, but controls for covariates (continuous variables that may affect the outcome).

- Adjusts the dependent variable for the influence of the covariate.

**Q3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.**

```python
def bernoulli_pdf(x, p):
    """
    Calculate the PDF of a Bernoulli distribution.

    Parameters:
    x : int (0 or 1) - the outcome
    p : float - the probability of success (P(X=1))

    Returns:
    float - the probability of the outcome x
    """
    if x not in [0, 1]:
        raise ValueError("x must be 0 or 1")
    if not (0 <= p <= 1):
        raise ValueError("p must be between 0 and 1")

    return p if x == 1 else 1 - p

# Example usage
print("P(X=0) when p=0.7:", bernoulli_pdf(0, 0.7))
print("P(X=1) when p=0.7:", bernoulli_pdf(1, 0.7))
```

**Q6. Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete uniform distribution.**

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
low = 1      # inclusive minimum value
high = 6     # inclusive maximum value (e.g. a fair 6‚Äësided die)

# Support values
values = np.arange(low, high + 1)

# PDF for discrete uniform: each value has equal probability
prob = 1 / (high - low + 1)
pdf = np.ones_like(values, dtype=float) * prob

# CDF: cumulative sum of PDF
cdf = np.cumsum(pdf)

# Display PDF and CDF values
print("Value | PDF      | CDF")
print("------------------------")
for v, p, c in zip(values, pdf, cdf):
    print(f"{v:5d} | {p:8.4f} | {c:8.4f}")

# Plotting the CDF
plt.step(values, cdf, where='post', label='CDF', linewidth=2)
plt.scatter(values, cdf, color='red')
plt.title(f"CDF of Discrete Uniform Distribution [{low}, {high}]")
plt.xlabel("Value")
plt.ylabel("Cumulative Probability")
plt.xticks(values)
plt.ylim(0, 1.05)
plt.grid(True)
plt.legend()
plt.show()
```

**Q9. Write a Python function to calculate Z-scores from a dataset and plot them.**

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_z_scores(data):
    """
    Calculates Z-scores of the given data and plots them.

    Parameters:
    data (list or array-like): Input numeric data
    """
    data = np.array(data)
    mean = np.mean(data)
    std = np.std(data)

    if std == 0:
        raise ValueError("Standard deviation is zero; Z-scores are undefined.")

    z_scores = (data - mean) / std

    # Plot Z-scores
    plt.scatter(range(len(z_scores)), z_scores, color='blue')
    plt.axhline(0, color='red', linestyle='--', label='Mean (Z=0)')
    plt.axhline(1, color='green', linestyle='--', label='Z=1')
    plt.axhline(-1, color='green', linestyle='--', label='Z=-1')
    plt.title("Z-scores of Dataset")
    plt.xlabel("Data Index")
    plt.ylabel("Z-score")
    plt.grid(True)
    plt.legend()
    plt.show()

    return z_scores

# Example usage:
data = [10, 12, 14, 16, 18, 20, 22]
z_scores = calculate_and_plot_z_scores(data)
print("Z-scores:", z_scores)
```

**Q14. Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution.**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_z_score(data, value):
    """
    Calculates the Z-score of a value relative to the dataset.

    Parameters:
    data : list or array-like of numeric values
    value : the data point to compute Z-score for

    Returns:
    z : Z-score of the given value
    """
    mean = np.mean(data)
    std = np.std(data)

    if std == 0:
        raise ValueError("Standard deviation is zero; Z-score is undefined.")

    z = (value - mean) / std
    return z

def plot_standard_normal_with_z(z):
    """
    Plots the standard normal distribution and highlights the given Z-score.
    """
    x = np.linspace(-4, 4, 1000)
    y = norm.pdf(x, 0, 1)

    plt.plot(x, y, label='Standard Normal Distribution', color='blue')
    plt.fill_between(x, 0, y, where=(x <= z), color='skyblue', alpha=0.5, label=f'Area ‚â§ Z={z:.2f}')
    plt.axvline(z, color='red', linestyle='--', label=f'Z = {z:.2f}')
    plt.title("Standard Normal Distribution with Z-score")
    plt.xlabel("Z")
    plt.ylabel("Density")
    plt.legend()
    plt.grid(True)
    plt.show()

# Example dataset and value
data = [12, 15, 14, 10, 18, 17, 11, 13, 15]
value = 17

# Calculate Z-score
z_score = calculate_z_score(data, value)
print(f"Z-score for value {value} is: {z_score:.2f}")

# Plot the Z-score on the standard normal distribution
plot_standard_normal_with_z(z_score)
```

**Q15. Implement hypothesis testing using Z-statistics for a sample dataset.**

```python
import numpy as np
from scipy.stats import norm

def z_test(sample_data, pop_mean, pop_std, alpha=0.05, alternative='two-sided'):
    """
    Performs a one-sample Z-test.

    Parameters:
        sample_data (list/array): The sample data.
        pop_mean (float): The population mean (null hypothesis).
        pop_std (float): The population standard deviation (assumed known).
        alpha (float): Significance level (default: 0.05).
        alternative (str): Type of test: 'two-sided', 'less', or 'greater'.

    Returns:
        z_stat (float): Z-test statistic
        p_value (float): P-value of the test
        decision (str): 'Reject H0' or 'Fail to reject H0'
    """
    sample_data = np.array(sample_data)
    sample_mean = np.mean(sample_data)
    n = len(sample_data)

    # Compute Z-statistic
    z_stat = (sample_mean - pop_mean) / (pop_std / np.sqrt(n))

    # Compute p-value
    if alternative == 'two-sided':
        p_value = 2 * (1 - norm.cdf(abs(z_stat)))
    elif alternative == 'less':
        p_value = norm.cdf(z_stat)
    elif alternative == 'greater':
        p_value = 1 - norm.cdf(z_stat)
    else:
        raise ValueError("alternative must be 'two-sided', 'less', or 'greater'")

    # Decision
    decision = 'Reject H0' if p_value < alpha else 'Fail to reject H0'

    return z_stat, p_value, decision

# Example usage
sample = [105, 110, 102, 108, 107, 111, 106, 109, 108, 107]
pop_mean = 100     # claimed population mean
pop_std = 10       # known population standard deviation

z, p, decision = z_test(sample, pop_mean, pop_std, alpha=0.05, alternative='greater')

print(f"Z-statistic: {z:.3f}")
print(f"P-value: {p:.4f}")
print(f"Decision: {decision}")
```

**Q16. Create a confidence interval for a dataset using Python and interpret the result.**

```python
import numpy as np
from scipy import stats

def confidence_interval(data, confidence=0.95):
    """
    Computes the confidence interval for the mean of a dataset using t-distribution.

    Parameters:
        data (list or array): Sample data
        confidence (float): Confidence level (default 0.95)

    Returns:
        (lower_bound, upper_bound): Tuple with confidence interval
    """
    data = np.array(data)
    n = len(data)
    mean = np.mean(data)
    std_err = stats.sem(data)  # Standard error of the mean
    t_crit = stats.t.ppf((1 + confidence) / 2, df=n-1)

    margin_of_error = t_crit * std_err
    lower_bound = mean - margin_of_error
    upper_bound = mean + margin_of_error

    return (lower_bound, upper_bound), mean

# Example usage
sample = [102, 98, 100, 105, 101, 99, 103, 100, 97, 104]
(ci_lower, ci_upper), sample_mean = confidence_interval(sample, confidence=0.95)

print(f"Sample Mean: {sample_mean:.2f}")
print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
```

```python
import numpy as np
from scipy.stats import norm

# Known population parameters
population_mean = 100     # Œº‚ÇÄ
population_std = 15       # œÉ

# Sample data
sample_data = [105, 110, 98, 102, 108, 95, 100, 107, 101, 99]
sample_size = len(sample_data)
sample_mean = np.mean(sample_data)

# Calculate the Z-score
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

# Calculate the p-value (two-tailed)
p_value = 2 * (1 - norm.cdf(abs(z_score)))

# Display the results
print("Sample Mean:", round(sample_mean, 2))
print("Z-score:", round(z_score, 3))
print("P-value:", round(p_value, 4))

# Interpretation
alpha = 0.05  # significance level
if p_value < alpha:
    print("Result: Reject the null hypothesis (significant difference).")
else:
    print("Result: Fail to reject the null hypothesis (no significant difference).")
```

**Q3.  Implement a one-sample Z-test using Python to compare the sample mean with the population mean.**

```python
import numpy as np
from scipy.stats import norm

def one_sample_z_test(sample_data, population_mean, population_std, alpha=0.05):
    # Sample statistics
    sample_size = len(sample_data)
    sample_mean = np.mean(sample_data)

    # Z-score calculation
    z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

    # Two-tailed p-value calculation
    p_value = 2 * (1 - norm.cdf(abs(z_score)))

    # Output
    print("Sample Mean:", round(sample_mean, 2))
    print("Z-Score:", round(z_score, 3))
    print("P-Value:", round(p_value, 4))

    # Hypothesis decision
    if p_value < alpha:
        print("Result: Reject the null hypothesis (significant difference).")
    else:
        print("Result: Fail to reject the null hypothesis (no significant difference).")

# -------------------------------
# Example usage
# -------------------------------

# Simulated sample data
np.random.seed(0)
sample = np.random.normal(loc=52, scale=10, size=40)  # Sample from normal distribution

# Known population parameters
pop_mean = 50     # Œº‚ÇÄ
pop_std = 10      # œÉ

# Run the Z-test
one_sample_z_test(sample, pop_mean, pop_std)
```

**Q4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot.**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Simulate sample data
np.random.seed(1)
sample = np.random.normal(loc=53, scale=10, size=40)  # Sample mean > population mean

# Known population parameters
pop_mean = 50
pop_std = 10
alpha = 0.05  # significance level

# Compute Z-score
sample_mean = np.mean(sample)
sample_size = len(sample)
z_score = (sample_mean - pop_mean) / (pop_std / np.sqrt(sample_size))

# Compute critical Z-values for two-tailed test
z_critical = norm.ppf(1 - alpha/2)

# Compute p-value
p_value = 2 * (1 - norm.cdf(abs(z_score)))

# Display results
print("Sample Mean:", round(sample_mean, 2))
print("Z-score:", round(z_score, 3))
print("P-value:", round(p_value, 4))

# Decision
if abs(z_score) > z_critical:
    print("Result: Reject the null hypothesis (significant difference).")
else:
    print("Result: Fail to reject the null hypothesis (no significant difference).")

# -----------------------
# Visualization
# -----------------------

# Plot standard normal distribution
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label="Standard Normal Distribution", color='blue')

# Shade critical regions
plt.fill_between(x, y, where=(x <= -z_critical), color='red', alpha=0.4, label='Rejection Region (Left)')
plt.fill_between(x, y, where=(x >= z_critical), color='red', alpha=0.4, label='Rejection Region (Right)')

# Plot z-score line
plt.axvline(z_score, color='green', linestyle='--', label=f'Z-score = {z_score:.2f}')

# Add labels and legend
plt.title('Two-Tailed Z-Test Visualization')
plt.xlabel('Z-score')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

**Q5.  Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_type1_type2(pop_mean_null, pop_mean_alt, std_dev, n, alpha=0.05):
    """
    Visualizes Type I and Type II errors for hypothesis testing using Z-distribution.

    Parameters:
    - pop_mean_null: Mean under null hypothesis (H0)
    - pop_mean_alt: Mean under alternative hypothesis (H1)
    - std_dev: Population standard deviation (assumed known)
    - n: Sample size
    - alpha: Significance level (default is 0.05)
    """
    # Standard error
    se = std_dev / np.sqrt(n)

    # Critical z-value for two-tailed test
    z_critical = norm.ppf(1 - alpha / 2)

    # Critical values on X-axis under H0
    lower_crit = pop_mean_null - z_critical * se
    upper_crit = pop_mean_null + z_critical * se

    # x-axis range
    x = np.linspace(pop_mean_null - 4*se, pop_mean_alt + 4*se, 1000)

    # PDF under H0 and H1
    y_null = norm.pdf(x, loc=pop_mean_null, scale=se)
    y_alt = norm.pdf(x, loc=pop_mean_alt, scale=se)

    # Plot H0 and H1 distributions
    plt.figure(figsize=(12, 6))
    plt.plot(x, y_null, label='H0 Distribution', color='blue')
    plt.plot(x, y_alt, label='H1 Distribution', color='orange')

    # Shade Type I Error (Œ±): Reject H0 when it's true
    plt.fill_between(x, y_null, where=(x <= lower_crit), color='red', alpha=0.4, label='Type I Error (Œ±)')
    plt.fill_between(x, y_null, where=(x >= upper_crit), color='red', alpha=0.4)

    # Shade Type II Error (Œ≤): Fail to reject H0 when H1 is true
    plt.fill_between(x, y_alt, where=((x > lower_crit) & (x < upper_crit)), color='purple', alpha=0.3, label='Type II Error (Œ≤)')

    # Annotations
    plt.axvline(lower_crit, color='black', linestyle='--', label='Critical Values')
    plt.axvline(upper_crit, color='black', linestyle='--')

    plt.title("Visualization of Type I and Type II Errors")
    plt.xlabel("Sample Mean")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid(True)
    plt.show()

    # Compute Type II Error probability (Œ≤)
    beta = norm.cdf(upper_crit, loc=pop_mean_alt, scale=se) - norm.cdf(lower_crit, loc=pop_mean_alt, scale=se)
    power = 1 - beta

    print(f"Type I Error (Œ±): {alpha:.2f}")
    print(f"Type II Error (Œ≤): {beta:.4f}")
    print(f"Power of the test (1 - Œ≤): {power:.4f}")
```

**Q6. Write a Python program to perform an independent T-test and interpret the results.**

```python
import numpy as np
from scipy.stats import ttest_ind

# Simulate two independent samples
np.random.seed(42)
group1 = np.random.normal(loc=70, scale=10, size=30)  # e.g., control group
group2 = np.random.normal(loc=75, scale=12, size=30)  # e.g., treatment group

# Perform Independent T-Test
t_stat, p_value = ttest_ind(group1, group2, equal_var=False)  # Welch‚Äôs t-test (no assumption of equal variances)

# Print results
print("Group 1 Mean:", round(np.mean(group1), 2))
print("Group 2 Mean:", round(np.mean(group2), 2))
print("T-statistic:", round(t_stat, 3))
print("P-value:", round(p_value, 4))

# Interpret the result
alpha = 0.05
if p_value < alpha:
    print("Result: Reject the null hypothesis (significant difference between group means).")
else:
    print("Result: Fail to reject the null hypothesis (no significant difference between group means).")
```

**Q7.  Perform a paired sample T-test using Python and visualize the comparison results.**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import ttest_rel

# Simulate paired sample data (e.g., before and after treatment)
np.random.seed(0)
before = np.random.normal(loc=70, scale=8, size=25)
after = before + np.random.normal(loc=3, scale=5, size=25)  # slight increase after treatment

# Perform paired t-test
t_stat, p_value = ttest_rel(before, after)

# Print results
print("Mean (Before):", round(np.mean(before), 2))
print("Mean (After):", round(np.mean(after), 2))
print("T-statistic:", round(t_stat, 3))
print("P-value:", round(p_value, 4))

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Result: Reject the null hypothesis (significant change between before and after).")
else:
    print("Result: Fail to reject the null hypothesis (no significant change between before and after).")

# -------------------------------
# Visualization
# -------------------------------

# Line plot: individual differences
plt.figure(figsize=(12, 5))

# Subplot 1: Line plot
plt.subplot(1, 2, 1)
for i in range(len(before)):
    plt.plot(['Before', 'After'], [before[i], after[i]], marker='o', color='gray', alpha=0.6)
plt.title("Paired Sample Comparison (Line Plot)")
plt.ylabel("Scores")

# Subplot 2: Bar plot of means
plt.subplot(1, 2, 2)
means = [np.mean(before), np.mean(after)]
plt.bar(['Before', 'After'], means, color=['blue', 'orange'], alpha=0.7)
plt.title("Mean Scores Before vs After")
plt.ylabel("Mean Score")

plt.tight_layout()
plt.show()
```

**Q8. Simulate data and perform both Z-test and T-test, then compare the results using Python.**

```python
import numpy as np
from scipy.stats import ttest_1samp, norm

# ------------------------
# Simulate sample data
# ------------------------
np.random.seed(0)
sample_size = 30
true_mean = 50           # Population mean under H0
sample = np.random.normal(loc=52, scale=10, size=sample_size)  # Simulated sample data

# ------------------------
# Z-test (known population std dev)
# ------------------------
pop_std = 10  # Assume known
sample_mean = np.mean(sample)
se_z = pop_std / np.sqrt(sample_size)
z_score = (sample_mean - true_mean) / se_z
p_value_z = 2 * (1 - norm.cdf(abs(z_score)))

# ------------------------
# T-test (unknown population std dev)
# ------------------------
t_stat, p_value_t = ttest_1samp(sample, true_mean)

# ------------------------
# Results
# ------------------------
print("Sample Size:", sample_size)
print("Sample Mean:", round(sample_mean, 2))
print("\n--- Z-Test ---")
print("Z-score:", round(z_score, 3))
print("Z-test P-value:", round(p_value_z, 4))

print("\n--- T-Test ---")
print("T-statistic:", round(t_stat, 3))
print("T-test P-value:", round(p_value_t, 4))

# ------------------------
# Interpretation
# ------------------------
alpha = 0.05
print("\nInterpretation (Œ± = 0.05):")
if p_value_z < alpha:
    print("Z-test: Reject the null hypothesis (significant difference).")
else:
    print("Z-test: Fail to reject the null hypothesis.")

if p_value_t < alpha:
    print("T-test: Reject the null hypothesis (significant difference).")
else:
    print("T-test: Fail to reject the null hypothesis.")
```

**Q9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.**

```python
import numpy as np
from scipy.stats import t

def confidence_interval_mean(data, confidence=0.95):
    """
    Calculates the confidence interval for the mean of a sample.

    Parameters:
    - data (list or np.array): The sample data
    - confidence (float): Confidence level (default is 0.95 for 95%)

    Returns:
    - (mean, lower_bound, upper_bound): Tuple containing the sample mean and CI bounds
    """
    n = len(data)
    mean = np.mean(data)
    std_err = np.std(data, ddof=1) / np.sqrt(n)  # Standard error using sample std dev
    t_crit = t.ppf((1 + confidence) / 2, df=n - 1)  # Two-tailed t-critical value
    margin_of_error = t_crit * std_err
    lower = mean - margin_of_error
    upper = mean + margin_of_error
    return mean, lower, upper
```

**Q10. Write a Python program to calculate the margin of error for a given confidence level using sample data.**

```python
import numpy as np
from scipy.stats import t

def calculate_margin_of_error(data, confidence=0.95):
    """
    Calculates the margin of error for a sample mean using the t-distribution.

    Parameters:
    - data (list or np.array): Sample data
    - confidence (float): Confidence level (default is 0.95 for 95%)

    Returns:
    - margin_of_error (float)
    """
    n = len(data)
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)
    standard_error = sample_std / np.sqrt(n)

    # t critical value for given confidence and degrees of freedom
    t_crit = t.ppf((1 + confidence) / 2, df=n - 1)

    margin_of_error = t_crit * standard_error
    return margin_of_error, sample_mean
```

**Q11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.**

```python
def bayes_theorem(prior, likelihood, marginal):
    """
    Calculate posterior probability using Bayes' Theorem.

    Parameters:
    - prior (P(H)): Prior probability of the hypothesis
    - likelihood (P(E|H)): Likelihood of evidence given hypothesis
    - marginal (P(E)): Total probability of the evidence

    Returns:
    - posterior (P(H|E))
    """
    posterior = (likelihood * prior) / marginal
    return posterior

# Given:
P_D = 0.01               # Prior: Probability of having the disease
P_not_D = 1 - P_D        # Complement: Not having the disease

P_Pos_given_D = 0.99     # Likelihood: True positive
P_Pos_given_not_D = 0.05 # False positive

# Total probability of testing positive (marginal)
P_Pos = (P_Pos_given_D * P_D) + (P_Pos_given_not_D * P_not_D)

# Posterior: P(Disease | Positive Test)
P_D_given_Pos = bayes_theorem(P_D, P_Pos_given_D, P_Pos)

# Display results
print(f"Probability the patient has the disease given a positive test: {P_D_given_Pos:.4f}")
```

**Q12. Perform a Chi-square test for independence between two categorical variables in Python.**

```python
import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency
import seaborn as sns
import matplotlib.pyplot as plt

# Simulated categorical data
data = {
    'Gender':     ['Male', 'Female', 'Female', 'Male', 'Female', 'Male', 'Male', 'Female', 'Female', 'Male',
                   'Female', 'Female', 'Male', 'Female', 'Male', 'Male', 'Female', 'Male', 'Female', 'Male'],
    'Preference': ['Tea', 'Coffee', 'Tea', 'Tea', 'Coffee', 'Tea', 'Coffee', 'Tea', 'Coffee', 'Coffee',
                   'Tea', 'Tea', 'Coffee', 'Tea', 'Coffee', 'Coffee', 'Tea', 'Tea', 'Coffee', 'Coffee']
}

df = pd.DataFrame(data)

# Create a contingency table
contingency_table = pd.crosstab(df['Gender'], df['Preference'])

print("Contingency Table:")
print(contingency_table)

# Perform Chi-square test
chi2_stat, p_val, dof, expected = chi2_contingency(contingency_table)

# Print results
print("\nChi-square Test Results:")
print(f"Chi-square Statistic: {chi2_stat:.3f}")
print(f"Degrees of Freedom: {dof}")
print(f"P-value: {p_val:.4f}")

# Interpretation
alpha = 0.05
if p_val < alpha:
    print("Conclusion: Reject the null hypothesis (variables are dependent).")
else:
    print("Conclusion: Fail to reject the null hypothesis (variables are independent).")

# Visualize contingency table
sns.heatmap(contingency_table, annot=True, cmap="YlGnBu", fmt="d")
plt.title("Contingency Table: Gender vs Preference")
plt.ylabel("Gender")
plt.xlabel("Preference")
plt.tight_layout()
plt.show()
```

**Q14. Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution.**

```python
import numpy as np
from scipy.stats import chisquare

# Observed frequencies from rolling a 6-sided die
observed = np.array([8, 9, 10, 11, 12, 10])

# Expected frequencies (uniform distribution)
expected = np.full_like(observed, fill_value=np.mean(observed))

# Perform Chi-square goodness-of-fit test
chi2_stat, p_val = chisquare(f_obs=observed, f_exp=expected)

# Output results
print("Observed:", observed)
print("Expected:", expected)
print(f"\nChi-square Statistic: {chi2_stat:.4f}")
print(f"P-value: {p_val:.4f}")

# Interpretation
alpha = 0.05
if p_val < alpha:
    print("Conclusion: Reject the null hypothesis (data does not fit expected distribution).")
else:
    print("Conclusion: Fail to reject the null hypothesis (data fits expected distribution).")
```
